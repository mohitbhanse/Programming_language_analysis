```python
df.DATE = pd.to_datetime(df.DATE)
```

```python
reshaped_df = df.pivot(index='DATE',columns='TAG',values='POSTS')
reshaped_df
```

```python
reshaped_df.fillna(0,inplace=True)
```

```python
reshaped_df.head()
```

```python
reshaped_df.isna().sum()
```

```python
plt.figure(figsize=(20,10))
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('years',fontsize=25)
plt.ylabel('number of posts',fontsize=20)
plt.ylim(0,35000)
plt.plot(reshaped_df.index, reshaped_df.java)
plt.plot(reshaped_df.index, reshaped_df.python)
```

```python
plt.figure(figsize=(20,10))
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('years',fontsize=25)
plt.ylabel('number of posts',fontsize=20)
plt.ylim(0,35000)

for columns in reshaped_df.columns:
    plt.plot(reshaped_df.index,reshaped_df[columns],
            linewidth=3, label=reshaped_df[columns].name)
plt.legend(fontsize=16)
```

# Looks like Python is the most popular programming language judging by the number of posts on Stack Overflow! Python for the win! =) 

```python
roll_df = reshaped_df.rolling(window=3).mean()
plt.figure(figsize=(16,10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Posts', fontsize=14)
plt.ylim(0, 35000)
 
# plot the roll_df instead
for column in roll_df.columns:
    plt.plot(roll_df.index, roll_df[column], 
             linewidth=3, label=roll_df[column].name)
 
plt.legend(fontsize=16)
```
